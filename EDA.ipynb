```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cleaned-Data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
events = ['100m', '200m', '400m', '800m', '1500m', '5000m', '10000m', 'Triple Jump', 'High Jump', 'Long Jump', 'Shot Put', 'Discus', 'Javelin']  # Add all the events you want to include

plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

df = df[df['Year'] != 2020]
df = df[df['Year'] != 2024]


for event in events:
    filtered_df_event = df[(df['Sex'] == 'Men') & (df['Division'] == 'World') & (df['Event'] == event)]
    average_points_event = filtered_df_event.groupby('Year')['Points'].mean()
    plt.plot(average_points_event.index, average_points_event.values, label=event)

plt.xlabel('Year')
plt.ylabel('Average Points')
plt.title('Average Points for Men, Kansas 6A, and Various Events')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

divisions = ['World', 'NCAA D-I', 'NCAA D-II', 'NCAA D-III', 'NAIA']

for division in divisions:
    filtered_df_division = df[(df['Sex'] == 'Men') & (df['Division'] == division)]
    average_points_division = filtered_df_division.groupby('Year')['Points'].mean()
    plt.plot(average_points_division.index, average_points_division.values, label=division)

plt.xlabel('Year')
plt.ylabel('Average Points')
plt.title('Average Points for Men, All Divisions, 800m')
plt.legend()
plt.show()
```
