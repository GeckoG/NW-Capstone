```python
events = ['100m', '200m', '400m', '800m', '1500m', '5000m', '10000m', 'Triple Jump', 'High Jump', 'Long Jump', 'Shot Put', 'Discus', 'Javelin']  # Add the events to include

plt.figure(figsize=(10, 6))  # Adjust the figure size if needed


for event in events:
    filtered_df_event = df[(df['Sex'] == 'Women') & (df['Division'] == 'NCAA D-II') & (df['Event'] == event)]
    average_points_event = filtered_df_event.groupby('Year')['Points'].mean()
    plt.plot(average_points_event.index, average_points_event.values, label=event)

plt.xlabel('Year')
plt.ylabel('Average Points')
plt.title('Average Points for Women, NCAA D-II, and Various Events')
plt.legend()
plt.show()
```

```python
filtered_df = df[(df['Division'] == 'NCAA D-I')]

# Convert 'Date' to datetime
filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])

# Create a crosstab heatmap
heatmap_data = pd.crosstab(filtered_df['Date'].dt.month, filtered_df['Date'].dt.day)

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='Blues')
plt.title('Heatmap of Performance Frequency by Month and Day')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

This is a disappointing discovery. It was anomalous to see NCAA athletes not setting personal bests at the end of May and into June. However, it seems that TFRRS does not include the NCAA Championships in their performance lists. While unfortunate, it is at least uniform and complete among the collegiate entries.

```python
df_filtered = df[(df['Event'] == '400m') & (df['Sex'] == 'Women') & (df['Division'] == 'NCAA D-I')]

# Calculate maximum and minimum marks for each year
max_marks = df_filtered.groupby('Year')['Mark'].max()
min_marks = df_filtered.groupby('Year')['Mark'].min()
average_marks = df_filtered.groupby('Year')['Mark'].mean()

# Plotting
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

# Plot top and bottom lines
plt.plot(max_marks.index, max_marks.values, color='gray', alpha=0.3, linestyle='-', label='Top Mark')
plt.plot(min_marks.index, min_marks.values, color='gray', alpha=0.3, linestyle='-', label='#100 Mark')

# Fill area between top and bottom lines with dark grey shade
plt.fill_between(max_marks.index, min_marks.values, max_marks.values, color='darkgray', alpha=0.5)

# Plot average line
plt.plot(average_marks.index, average_marks.values, color='black', linestyle='-', label='Average of top 100')

plt.xlabel('Year')
plt.ylabel('Mark')
plt.title("Top and Bottom Marks by Year (D1 Women's 400m)")
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()  # Invert y-axis
plt.tight_layout()
plt.show()
```

This shows both the top mark and 100th mark in relation to the average.

```python
df_filtered = df[(df['Event'] == '1500m') & (df['Sex'] == 'Women') & (df['Division'] == 'NCAA D-I')]
average_marks = df_filtered.groupby('Year')['Mark'].mean()

plt.figure(figsize=(12, 6))  # Adjust figure size if needed

# Loop through each year and plot marks vertically
for year, group in df_filtered.groupby('Year'):
    marks = group['Mark']
    plt.eventplot(marks, orientation='vertical', lineoffsets=year, colors='black', linewidths=0.5)

plt.plot(average_marks.index, average_marks.values, color='red', linestyle='-', label='Average of top 100')
plt.xlabel('Year')
plt.ylabel('Performance (seconds)')
plt.title("NCAA D-I Women's 1500m Marks by Year")
plt.grid(axis='y')
plt.gca().invert_yaxis()  # Invert y-axis to have the years in ascending order
plt.tight_layout()
plt.show()
```

This is a better visualization, since it shows where each mark lies.
